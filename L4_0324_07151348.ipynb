```python
import seaborn as sns
```

**下載seaborn內建的資料集tips，這個資料集紀錄了餐廳客人給得小費以及相關的基本資料。**

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.groupby(['smoker','sex']).size().unstack(1).plot(kind='bar')
#用tips的資料分群，分成有吸菸沒吸菸 男生女生，size是把人數加在一起(吸菸的男生有多少 沒吸菸的男生有多少)，unstack(1)是要求不要疊加起來
```

```python
tips.groupby('sex')['total_bill'].plot(kind='hist', bins=30, alpha=0.5, legend=True) #alpha 是透明度的參數設定。
```

```python
tips.boxplot(column='total_bill',by='sex')
```

**Scatter Plot**散佈圖

```python
df1.plot.scatter(x='A',y='B') #裡面是兩個連續變數
```

```python
# color based off another column value資料分布和上圖一樣 但差別在用c的大小來區分它的顏色 
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
```

```python
df1.plot.scatter(x='A',y='B',s=df1['C']*200)#上圖是用c的顏色分 這裡適用c的大小
```

**HeatMap**

```python
# Matrix form for correlation data看資料的關聯性 相關係數   若相關係數低 可以看要不要整合他或刪除他
tips.corr()
```

**pairplot**

**lmplot**

```python
plt.plot(x,y,label='y = x ** 3')
plt.legend()
```

```python
z=pow(x,4)
w=pow(x,5)
plt.figure(figsize=(8,6))#圖的大小要八乘六
plt.plot(x,y)
plt.plot(x,z)
plt.plot(x,w)
plt.title("The result")
plt.text(4.5,200,'x**3')
plt.text(4,400,'x**4')
plt.text(3.5,900,'x**5')
plt.show()
```
